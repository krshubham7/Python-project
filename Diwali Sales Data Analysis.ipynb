# **Data Analysis**

**1. Gender :**

```python
df.columns
```

**1.1 Gender Vs Order Count Plot :**

```python
ax = sns.countplot(x = 'Gender', data = df, hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

**1.2 Gender Vs Amount Plot :**

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Gender', y='Amount', data = sales_gen, hue = 'Gender')
```

**From the above graphs, we can see that most of the buyers are females and even the purchasing power of females are greater than males**

**2. Age Group :**

```python
df.columns
```

**2.1 Age Group Vs Order Count Plot :**

```python
cx = sns.countplot(x= 'Age Group', data = df, hue = 'Gender')

for bars in cx.containers:
    cx.bar_label(bars)
```

From the above graph, we can see that female customer of age group 26-35 have maximum number of orders

**2.2 Age Group Vs Amount Plot :**

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

dx = sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age,hue = 'Age Group')
```

**Combining the above two graphs, we can see that max amount spent and maximum number of order is placed by female customers having age in between 26-35**

**3. State**

```python
df.columns
```

**3.1 State Vs Order count plot**

```python
sales_state1 = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head()

sns.barplot(x = 'State', y = 'Orders',data = sales_state1, palette = "Set2")
```

**3.2 Status Vs Amount Plot :**

```python


sales_state2 = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head()


sns.barplot(x = 'State', y = 'Amount', data = sales_state2, palette = "Set2")
```

**From the above two graphs, it is clear that the max number of orders and maximum amount spent- both are from Uttar Pradesh followed by Maharashtra and Karnataka**

**4. Marital Status :**

```python
df.columns
```

**4.1 Marital Status Vs Order count Plot :**

```python
ax = sns.countplot(x = 'Marital_Status', data = df, hue = 'Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

**4.2 Marital Status Vs Amount Plot:**

```python
sales_shadi = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_shadi,hue = 'Gender')
```

**From the above graph we can see that most of the buyers are married (women) and they have high purchasing power**

**5. Occupation**

**5.1 Occupation Vs Order count Plot :**

```python
ax = sns.countplot(x = 'Occupation', data = df, hue = 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

**5.2 Occupation Vs Amount Plot :**

```python
sales_occupation = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Occupation', y = 'Amount', data = sales_occupation, hue = 'Occupation')
```

**From the above two graphs, we can see that most of the buyers are working in IT,  Healthcare and Aviation sector**

**6. Product Category**

```python
df.columns
```

```python
sns.countplot(x = 'Product_Category', data = df, hue = 'Product_Category')
```

**6.2 Product Category Vs Amount Plot :**

```python
sales_productcat = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head()

sns.barplot(x= 'Product_Category', y = 'Amount', data = sales_productcat, hue ='Product_Category' )
```

From the above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
